```python
OPENAI_API_KEY = ""
openai.api_key = OPENAI_API_KEY
dataset_path = "/content/gdrive/MyDrive/Jacopo/dataset/test.json"
```

```python
def parafrasi(subset_of_statements, size):
  frase = "Parafrasa le seguenti frasi:"
  for i in range(size):
    frase += (" " + str(i+1)+") " + subset_of_statements[i])

  prompt = str(frase)
  response = openai.ChatCompletion.create(
    model = "gpt-3.5-turbo",
    messages = [{
              "role": "user",
              "content": prompt
          }],
    temperature = 1,
    n = 5,
    max_tokens = 2000
  )

  resp_content = []
  for choice in response.choices:
    resp_content.append(choice.message.content)
  return resp_content
```

Estrarre il testo della parafrasi da fare dal drive e buttare sempre nel drive il testo ottenuto

```python
dataset_file = open(dataset_path)
statements = json.load(dataset_file)
dataset_file.close()
```

```python
print(len(statements))
```

```python
statements_per_query = 2
responses = []
for i in range(194, 206, statements_per_query):
  try :
    responses.append(parafrasi(statements[i:i+statements_per_query] , statements_per_query))
  except:
    responses.append(parafrasi(statements[i:(len(statements))] , len(statements) - i))
```

```python
with open("/content/gdrive/MyDrive/Jacopo/exp_chatGPT/complete-2perquery6.json", "w") as F:
    json.dump(responses, F, indent=2, ensure_ascii=False)
```
