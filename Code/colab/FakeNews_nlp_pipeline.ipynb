```python
import stanza
stanza.download('it') # download italian model
nlp = stanza.Pipeline(lang='it', processors='tokenize')
```

```python
import json
```

```python
#data = json.load(uploaded.get("mini.json").decode("utf-8"))
content_file = "/content/gdrive/MyDrive/Colab Notebooks/fakenews_data/contents.json"
d1_file = "/content/gdrive/MyDrive/Colab Notebooks/fakenews_data/dataset_v1.json"
d2_file = "/content/gdrive/MyDrive/Colab Notebooks/fakenews_data/dataset_v2.json"
d3_file = "/content/gdrive/MyDrive/Colab Notebooks/fakenews_data/dataset_v3.json"
out_path = "/content/gdrive/MyDrive/Colab Notebooks/fakenews_data/out/"
```

```python
apiContent = open("/content/gdrive/MyDrive/Colab Notebooks/fakenews_data/mini.json")
data = json.load(apiContent)
apiContent.close()

for articolo in data:
  contenuto = articolo["content"]
  doc = nlp(contenuto) # run annotation over a sentence
  articolo["content"] = [sentence.text for sentence in doc.sentences]

with open("/content/gdrive/MyDrive/Colab Notebooks/fakenews_data/out/" + "mini.json", 'w') as json_file:
  json.dump(data, json_file, indent=2, ensure_ascii=False)
```
